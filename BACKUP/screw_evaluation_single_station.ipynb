### Filterung der allgemeinen Daten

```python
#select path
path = r"C:\Users\ukusp003\Desktop\Schraubdatenauswertung\PFExport_Station_1_2025-08-20T12_12_02_Results.csv"
#load raw data
df = pd.read_csv(path, sep = ";", usecols = [1, 6, 10, 11, 12],skiprows = 2, header = None)
#split cols with date and time and drop useless col with date+time
df[["Datum", "Uhrzeit"]] = df[6].str.split(" ", expand = True)
df = df.drop(6, axis = 1)
#evaluate a minimum start date to drop older screw data
df["Datum"] = pd.to_datetime(df["Datum"], format="%d.%m.%Y")
df = df[df["Datum"] >= pd.to_datetime("01.07.2025", format="%d.%m.%Y")]
#set column headers
header = ["Station", "Status", "Statusinfo", "Bauteil", "Datum", "Uhrzeit"]
df.columns = header
station = df.loc[0, "Station"]
```

### Filterung zur Analyse und Visualisierung

```python
#Filterung nach Datum
start_date = pd.to_datetime("2025-07-08")
end_date = pd.to_datetime("2025-07-08")
df_filtered = df[(df["Datum"] >= start_date) & (df["Datum"] <= end_date)]
```

```python
#Export detailed screw_data
df_grouped_detailed = df_filtered.groupby([df_filtered["Datum"].dt.date, "Station", "Bauteil", "Status"]).size().unstack(fill_value=0)
df_grouped_detailed.head()

#sort for most NIO
df_grouped_detailed = df_grouped_detailed.reset_index()
if "Verschraubung NIO" in df_grouped_detailed.columns:
    df_grouped_detailed = df_grouped_detailed.sort_values(
        by=["Datum", "Station", "Verschraubung NIO"], 
        ascending=[True, True, False]
    )
df_grouped_detailed = df_grouped_detailed.set_index(["Datum", "Station", "Bauteil"])

df_grouped_detailed.tail()
df_grouped_detailed.to_excel("df_grouped_detailed.xlsx")
```

```python
#groupby aggregation
df_grouped_super_detailed = df_filtered.groupby([df_filtered["Datum"].dt.date,"Station", "Bauteil", "Statusinfo"]).size().unstack()

#delete cols without entry
df_grouped_super_detailed = df_grouped_super_detailed.loc[:, (df_grouped_super_detailed != 0).any(axis=0)]

#fill up int values 
df_grouped_super_detailed = df_grouped_super_detailed.fillna(0).astype(int)

cols = df_grouped_super_detailed.columns.tolist()
if "Verschraubung IO" in cols: 
    cols = ["Verschraubung IO"] + [col for col in cols if col != "Verschraubung IO"] 
    df_grouped_super_detailed = df_grouped_super_detailed[cols]


# Filter auf Bauteile, die "FAT" enthalten
df_grouped_super_detailed = df_grouped_super_detailed.loc[df_grouped_super_detailed.index.get_level_values("Bauteil").str.contains("FAT", case=False, na=False)]

df_grouped_super_detailed.head()

df_grouped_super_detailed.to_excel("df_grouped_super_detailed.xlsx")
```

# PLOT

```python
#Filtering for plot
component = ["FAT L", "FAT R", "FOT-V L", "FOT-V R", "FOT-W L", "FOT-W R", "FOT-Z L", "FOT-Z R"]
df_filtered = df_filtered[df_filtered["Bauteil"].str.contains("|".join(component), na=False)].copy()
df_filtered["Bauteil"] = df_filtered["Bauteil"].apply(lambda x: next((b for b in component if b in x), x))
df_grouped_plot = df_filtered.groupby([df_filtered["Datum"].dt.date, "Bauteil", "Status"]).size().unstack(fill_value=0)
```

```python
#Create plot
df_grouped_plot = df_grouped_plot.reset_index()
```

```python
if "Verschraubung NIO" in df_grouped_plot.columns:
    pass
else:
    df_grouped_plot["Verschraubung NIO"] = 0
```

```python
df_grouped_plot_copy = df_grouped_plot.copy()
print(df_grouped_plot_copy)
df_grouped_plot_copy["Prozentualer Fehler"] = (df_grouped_plot_copy['Verschraubung NIO'] / (df_grouped_plot_copy['Verschraubung IO'] + df_grouped_plot_copy['Verschraubung NIO'])) * 100
#Filterung nach FAT
df_grouped_plot_fat = df_grouped_plot_copy[df_grouped_plot_copy["Bauteil"].str.contains("FAT", na=False)]
#Filterung nach FOT V
df_grouped_plot_fotv = df_grouped_plot_copy[df_grouped_plot_copy["Bauteil"].str.contains("FOT-V", na=False)]
#Filterung nach FOT W
df_grouped_plot_fotw= df_grouped_plot_copy[df_grouped_plot_copy["Bauteil"].str.contains("FOT-W", na=False)]
#Filterung nach FOT Z
df_grouped_plot_fotz = df_grouped_plot_copy[df_grouped_plot_copy["Bauteil"].str.contains("FOT-Z", na=False)]


#FAT
df_plot_io_nio = df_grouped_plot_fat[['Verschraubung IO', 'Verschraubung NIO']]
df_plot_fehler = df_grouped_plot_fat[['Prozentualer Fehler']]
# Prüfen, ob Daten vorhanden sind
has_io_nio = not df_plot_io_nio.dropna(how='all').empty and df_plot_io_nio.sum().sum() > 0
has_fehler = not df_plot_fehler.dropna(how='all').empty and df_plot_fehler.sum().sum() > 0
# Anzahl der Subplots bestimmen
num_plots = sum([has_io_nio, has_fehler])
if num_plots == 0:
    print("Keine Daten für IO/NIO oder Prozentualer Fehler vorhanden – Diagramm wird nicht erstellt.")
else:
    fig, axes = plt.subplots(num_plots, 1, figsize=(12, 6 * num_plots), sharex=True)
    # Falls nur ein Plot vorhanden ist, axes in Liste umwandeln
    if num_plots == 1:
        axes = [axes]
    plot_idx = 0
    if has_io_nio:
        df_grouped_plot_fat.set_index(['Datum', 'Bauteil'])[['Verschraubung IO', 'Verschraubung NIO']].plot(
            kind='bar', stacked=False, color=['darkgreen', 'red'], edgecolor='black', ax=axes[plot_idx]
        )
        axes[plot_idx].set_title(f'Verschraubungen IO vs NIO FAT, {station}\nZeitraum {start_date.date()} - {end_date.date()}',
                                 fontsize=12, fontweight='bold')
        axes[plot_idx].set_ylabel('Anzahl')
        axes[plot_idx].grid(color='gray', linewidth=0.25)
        plot_idx += 1

    if has_fehler:
        df_grouped_plot_fat.set_index(['Datum', 'Bauteil'])[['Prozentualer Fehler']].plot(
            kind='bar', stacked=False, color='royalblue', edgecolor='black', ax=axes[plot_idx]
        )
        axes[plot_idx].set_title(f'Prozentualer Fehler FAT, {station}\nZeitraum {start_date.date()} - {end_date.date()}',
                                 fontsize=12, fontweight='bold')
        axes[plot_idx].set_ylabel('Fehler in %')
        axes[plot_idx].set_ylim(0, 100)
        axes[plot_idx].grid(color='gray', linewidth=0.25)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

#FOT-V
df_plot_io_nio = df_grouped_plot_fotv[['Verschraubung IO', 'Verschraubung NIO']]
df_plot_fehler = df_grouped_plot_fotv[['Prozentualer Fehler']]
# Prüfen, ob Daten vorhanden sind
has_io_nio = not df_plot_io_nio.dropna(how='all').empty and df_plot_io_nio.sum().sum() > 0
has_fehler = not df_plot_fehler.dropna(how='all').empty and df_plot_fehler.sum().sum() > 0
# Anzahl der Subplots bestimmen
num_plots = sum([has_io_nio, has_fehler])
if num_plots == 0:
    print("Keine Daten für IO/NIO oder Prozentualer Fehler vorhanden – Diagramm wird nicht erstellt.")
else:
    fig, axes = plt.subplots(num_plots, 1, figsize=(12, 6 * num_plots), sharex=True)
    # Falls nur ein Plot vorhanden ist, axes in Liste umwandeln
    if num_plots == 1:
        axes = [axes]
    plot_idx = 0
    if has_io_nio:
        df_grouped_plot_fotv.set_index(['Datum', 'Bauteil'])[['Verschraubung IO', 'Verschraubung NIO']].plot(
            kind='bar', stacked=False, color=['darkgreen', 'red'], edgecolor='black', ax=axes[plot_idx]
        )
        axes[plot_idx].set_title(f'Verschraubungen IO vs NIO FOT-V, {station}\nZeitraum {start_date.date()} - {end_date.date()}',
                                 fontsize=12, fontweight='bold')
        axes[plot_idx].set_ylabel('Anzahl')
        axes[plot_idx].grid(color='gray', linewidth=0.25)
        plot_idx += 1

    if has_fehler:
        df_grouped_plot_fotv.set_index(['Datum', 'Bauteil'])[['Prozentualer Fehler']].plot(
            kind='bar', stacked=False, color='royalblue', edgecolor='black', ax=axes[plot_idx]
        )
        axes[plot_idx].set_title(f'Prozentualer Fehler FOT-V, {station}\nZeitraum {start_date.date()} - {end_date.date()}',
                                 fontsize=12, fontweight='bold')
        axes[plot_idx].set_ylabel('Fehler in %')
        axes[plot_idx].set_ylim(0, 100)
        axes[plot_idx].grid(color='gray', linewidth=0.25)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
 
#FOT-W
df_plot_io_nio = df_grouped_plot_fotw[['Verschraubung IO', 'Verschraubung NIO']]
df_plot_fehler = df_grouped_plot_fotw[['Prozentualer Fehler']]
# Prüfen, ob Daten vorhanden sind
has_io_nio = not df_plot_io_nio.dropna(how='all').empty and df_plot_io_nio.sum().sum() > 0
has_fehler = not df_plot_fehler.dropna(how='all').empty and df_plot_fehler.sum().sum() > 0
# Anzahl der Subplots bestimmen
num_plots = sum([has_io_nio, has_fehler])
if num_plots == 0:
    print("Keine Daten für IO/NIO oder Prozentualer Fehler vorhanden – Diagramm wird nicht erstellt.")
else:
    fig, axes = plt.subplots(num_plots, 1, figsize=(12, 6 * num_plots), sharex=True)
    # Falls nur ein Plot vorhanden ist, axes in Liste umwandeln
    if num_plots == 1:
        axes = [axes]
    plot_idx = 0
    if has_io_nio:
        df_grouped_plot_fotw.set_index(['Datum', 'Bauteil'])[['Verschraubung IO', 'Verschraubung NIO']].plot(
            kind='bar', stacked=False, color=['darkgreen', 'red'], edgecolor='black', ax=axes[plot_idx]
        )
        axes[plot_idx].set_title(f'Verschraubungen IO vs NIO FOT-W, {station}\nZeitraum {start_date.date()} - {end_date.date()}',
                                 fontsize=12, fontweight='bold')
        axes[plot_idx].set_ylabel('Anzahl')
        axes[plot_idx].grid(color='gray', linewidth=0.25)
        plot_idx += 1

    if has_fehler:
        df_grouped_plot_fotw.set_index(['Datum', 'Bauteil'])[['Prozentualer Fehler']].plot(
            kind='bar', stacked=False, color='royalblue', edgecolor='black', ax=axes[plot_idx]
        )
        axes[plot_idx].set_title(f'Prozentualer Fehler FOT-W, {station}\nZeitraum {start_date.date()} - {end_date.date()}',
                                 fontsize=12, fontweight='bold')
        axes[plot_idx].set_ylabel('Fehler in %')
        axes[plot_idx].set_ylim(0, 100)
        axes[plot_idx].grid(color='gray', linewidth=0.25)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

#FOT-Z
df_plot_io_nio = df_grouped_plot_fotz[['Verschraubung IO', 'Verschraubung NIO']]
df_plot_fehler = df_grouped_plot_fotz[['Prozentualer Fehler']]
# Prüfen, ob Daten vorhanden sind
has_io_nio = not df_plot_io_nio.dropna(how='all').empty and df_plot_io_nio.sum().sum() > 0
has_fehler = not df_plot_fehler.dropna(how='all').empty and df_plot_fehler.sum().sum() > 0
# Anzahl der Subplots bestimmen
num_plots = sum([has_io_nio, has_fehler])
if num_plots == 0:
    print("Keine Daten für IO/NIO oder Prozentualer Fehler vorhanden – Diagramm wird nicht erstellt.")
else:
    fig, axes = plt.subplots(num_plots, 1, figsize=(12, 6 * num_plots), sharex=True)
    # Falls nur ein Plot vorhanden ist, axes in Liste umwandeln
    if num_plots == 1:
        axes = [axes]
    plot_idx = 0
    if has_io_nio:
        df_grouped_plot_fotz.set_index(['Datum', 'Bauteil'])[['Verschraubung IO', 'Verschraubung NIO']].plot(
            kind='bar', stacked=False, color=['darkgreen', 'red'], edgecolor='black', ax=axes[plot_idx]
        )
        axes[plot_idx].set_title(f'Verschraubungen IO vs NIO FOT-Z, {station}\nZeitraum {start_date.date()} - {end_date.date()}',
                                 fontsize=12, fontweight='bold')
        axes[plot_idx].set_ylabel('Anzahl')
        axes[plot_idx].grid(color='gray', linewidth=0.25)
        plot_idx += 1

    if has_fehler:
        df_grouped_plot_fotz.set_index(['Datum', 'Bauteil'])[['Prozentualer Fehler']].plot(
            kind='bar', stacked=False, color='royalblue', edgecolor='black', ax=axes[plot_idx]
        )
        axes[plot_idx].set_title(f'Prozentualer Fehler FOT-Z, {station}\nZeitraum {start_date.date()} - {end_date.date()}',
                                 fontsize=12, fontweight='bold')
        axes[plot_idx].set_ylabel('Fehler in %')
        axes[plot_idx].set_ylim(0, 100)
        axes[plot_idx].grid(color='gray', linewidth=0.25)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```
